### Activity#1
1. Create a list of 100 random numbers.
2. Create a new list from this random list, with numbers that are divisible by 3.
3. Calculate the length of these two lists and store the difference in a new variable.
4. Using a loop, perform steps 2 and 3 and find the difference variable three times.
5. Find the arithmetic mean of these three difference values.

```python
# Create a list of 100 random numbers.
import random
my_rdm_list = random.sample(range(1, 150), 100)
print(my_rdm_list)
```

```python
# Create a new list from this random list, with numbers that are divisible by 3.
my_new_rdm_list = [i for i in my_rdm_list if i % 3 == 0]

print(my_new_rdm_list)
```

```python
# Calculate the length of these two lists and store the difference in a new variable.
len_my_rdm_list = len(my_rdm_list)
len_my_new_rdm_list = len(my_new_rdm_list)

len_diff = len_my_rdm_list - len_my_new_rdm_list
print(len_diff)
```

```python
# Using a loop, perform steps 2 and 3 and find the difference variable three times.

diff_list = []
for i in range(0, 3):
    my_rdm_list = random.sample(range(1, 150), 100)
    my_new_rdm_list = [i for i in my_rdm_list if i % 3 == 0]
    
    len_my_rdm_list = len(my_rdm_list)
    len_my_new_rdm_list = len(my_new_rdm_list)

    len_diff = len_my_rdm_list - len_my_new_rdm_list
    diff_list.append(len_diff)
diff_list
```

```python
# Find the arithmetic mean of these three difference values.
avg_diff = sum(diff_list) / float(len(diff_list))
print(avg_diff)
```

### Activity 4: Design Your Own CSV Parser

1. Import zip_longest from itertools. Create a function to zip header, line and fillvalue=None.
2. Open the accompanying sales_record.csv file from the GitHub link by using r mode inside a with block and first check that it is opened.
3. Read the first line and use string methods to generate a list of all the column names.
4. Start reading the file. Read it line by line.
5. Read each line and pass that line to a function, along with the list of the headers. The work of the function is to construct a dict out of these two and fill up the key:values. Keep in mind that a missing value should result in None.

```python
from itertools import zip_longest
```

```python
def return_dict_from_csv_line(header, line):
    # Zip them
    zipped_line = zip_longest(header, line, fillvalue=None)
    # Use dict comprehension to generate the final dict
    ret_dict = {kv[0]: kv[1] for kv in zipped_line}
    return ret_dict
```

```python
with open("/Users/siddharthabhaumik/Downloads/sales_record.csv", "r") as fd:
    first_line = fd.readline()
    header = first_line.replace("\n", "").split(",")
    for i, line in enumerate(fd):
        # Sample 10 rows
        line = line.replace("\n", "").split(",")
        d = return_dict_from_csv_line(header, line)
        print(d)
        if i > 10:
            break
```
